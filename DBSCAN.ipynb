```python
import pandas as pd
import numpy as np
df=pd.read_csv('target.csv',encoding = "ISO-8859-1", engine='python')
df.shape

```

```python
df.columns
df.isnull().sum()
df=df.dropna(axis='columns')
df[['Name','Address.City']]
```

```python
df_tar= df[['Address.City','Address.Latitude','Address.Longitude']]
df_tar=df_tar.set_index('Address.City')
df_tar
```

```python
# Convert kilometers to radians
# UPDATE.
# km radians = distance in km / 6371.
# mi radians = distance in mi / 3959.
# To convert:
# distance to radians: divide the distance by the radius of the sphere (e.g. the Earth) in the same units as the distance measurement.

#convert 300 miles into radian which will be epsilon val
eps=300/3959
eps
```

```python
from sklearn.cluster import DBSCAN
db_scan=DBSCAN(eps=eps,min_samples=3,algorithm='auto',metric='euclidean').fit(np.radians(df_tar))
clustr_labels=db_scan.labels_
clustr_labels
num_cluster=len(set(clustr_labels))

num_cluster
```

```python
df['DB_Cluster']=clustr_labels
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df['Address.Latitude'], df['Address.Longitude'])
```

```python
import seaborn as sns; sns.set()
plt.figure(figsize=(15,10))
ax=sns.scatterplot(x=df['Address.Latitude'],
                   y=df['Address.Longitude'],
                   hue=df['DB_Cluster'],
                   data=df)
plt.xlabel('Target Address Latitude')
plt.ylabel('Target Address Longitude')
```
